```python
# Testing out http://stackoverflow.com/questions/6620471/fitting-empirical-distribution-to-theoretical-ones-with-scipy-python
import matplotlib.pyplot as plt
import scipy
import scipy.stats
size = 30000
x = scipy.arange(size)
x
```

```python
y = scipy.int_(scipy.round_(scipy.stats.vonmises.rvs(5,size=size)*47))
y
```

```python
h = plt.hist(y, bins=range(48), color='w')
plt.show()
```

```python
from scipy.stats import gumbel_r
param = gumbel_r.fit(y)
param
```

```python
pdf_fitted = gumbel_r.pdf(x, *param[:-2], loc=param[-2], scale=param[-1]) * size
plt.plot(pdf_fitted, label="Gumbel R")
plt.hist(y, bins=range(48), color='w')
plt.xlim(0,47)
plt.legend(loc='upper right')
plt.show()
```
